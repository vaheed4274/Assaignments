```python
!pip install nltk
import io
import nltk
```

```python
f = io.open('sampletext.txt',mode='r',encoding='utf-8')
doc = f.read()
doc
```

# **Step 1: Sentence segmentation.**

```python
from nltk import sent_tokenize
sentences = sent_tokenize(doc)
sentences
```

# **Step 2: Word Tokenization**

```python
from nltk.tokenize import word_tokenize
words = []
for line in sentences:
  words.append(word_tokenize(line))
words
```

# **Step 3: Removal of Stop words**

```python
from nltk.corpus import stopwords
word_list = []
for line in words:
  for word in line:
    if word not in stopwords.words('english'):
      word_list.append(word)
print(len(word_list))
word_list
```

# **Step 4: Stemming & Lemmatization**

```python
from nltk.stem import PorterStemmer
ps =PorterStemmer()
stem_words = []
for w in word_list:
    rootword=ps.stem(w)
    stem_words.append(rootword)
stem_words
```

```python
from nltk.stem import 	WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
lemma_words = []
for w in word_list:
  lemma_words.append(wordnet_lemmatizer.lemmatize(w))
  print("Lemma for {} is {}".format(w, wordnet_lemmatizer.lemmatize(w)))
  
```

# **Step 5: Parts of Speech identification**

```python
tagged = nltk.pos_tag(lemma_words)
tagged
```

# **Step 6: Name entity and Facts**

```python
!pip install textacy
!python -m spacy download en_core_web_sm
import spacy
import textacy
```

```python
nlp=spacy.load('en_core_web_sm')
dc = nlp(doc)
for entity in dc.ents:
    print(f"{entity.text} ({entity.label_})")
```

```python
statements = textacy.extract.semistructured_statements(dc, entity="Hyderabad",cue='be')
print("All i know about Hyderabad:")
for statement in statements:
    subject, verb, fact = statement
    print(f" - {fact}")
```
